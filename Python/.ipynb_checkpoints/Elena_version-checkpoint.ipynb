### Итоговое задание:

f(x) = -12*x\**4*sin(cos(x)) - 18*x\**3+5*x\**2 + 10*x - 30

Определить корни

Найти интервалы, на которых функция возрастает

Найти интервалы, на которых функция убывает

Построить график

Вычислить вершину

Определить промежутки, на котором f > 0

Определить промежутки, на котором f < 0

```python
import matplotlib.pyplot as plt
from scipy.optimize import fsolve
import numpy as np
import sympy
```

```python
def f(x):
    return  -12*(x**4)*np.sin(np.cos(x))-18*x**3+5*x**2+10*x-30
```

```python
# Построим график функции на интервале [-100, 100]
x = np.arange(-100, 100, 1.5)
plt.plot(x, f(x), 'r-')
plt.grid()
```

```python
# Из графика функции очевидно, что функция имеет бесконечное множество корней и поведение функции повторяется
# (возрастает - убывает - возрастает - убывает - ...). Таким образом функцию необходимо рассматривать интервалами.
# Построим график функции на интервале [-2.5, 2.5]
x = np.arange(-2.5, 2.5, 0.01)
plt.plot(x, f(x), 'g-')
plt.grid()
```

```python
# Из графика очевидно, что функция равна нулю, в двух точках, найдем значение х и найдем вершину (min_y) на данном интервале

# Корни уравнения:
x1 = fsolve(f, -2)
x2 = fsolve(f, 2)
roots = [*x1, *x2]
roots = [round(x, 3) for x in roots]
x = np.arange(-2.5, 2.5, 0.01)
plt.plot(x1, f(x1), 'ro', label=f'Корни уравнения: {roots}')
plt.plot(x2, f(x2), 'ro')


# Вершина уравнения - минимальное значение функции на данном интервале:

min_y = min(f(x))

def y(x, res_y):
    return -12*(x**4)*np.sin(np.cos(x))-18*x**3+5*x**2+10*x-30-res_y

min_x = fsolve(y,1, min_y)
plt.plot(min_x, f(min_x), 'rx', label=f'Вершина: ({round(min_x[0], 3)}, {round(min_y, 3)} )')

# Функция f>0 и f<0:
x = np.arange(-2.5, x1, 0.01)
plt.plot(x, f(x), 'y-', label='Функция > 0')
x = np.arange(x1, x2, 0.01)
plt.plot(x, f(x), 'g-', label='Функция < 0')
x = np.arange(x2, 2.5, 0.01)
plt.plot(x, f(x), 'y-')

plt.axhline(0,color='blue')
plt.axvline(0,color='blue')
plt.legend()
plt.grid()

print(f'Поведение функции на интервале от {roots[0]} до {roots[1]} рассмотрим отдельно.\nНа интервале от -2.5 до {roots[0]} функция убывает, от {roots[1]} до 2.5 возрастает. На этих интервалах функция также > 0')
```

```python
# Поведение функции на участке от x1 до x2 (корни уравнения): 

x = np.arange(-1, 0.5, 0.01)
min_y1 = min(f(x))
a1 = fsolve(y, -1, min_y1)

x = np.arange(0, 1, 0.01)
min_y2 = max(f(x))
a2 = fsolve(y, 0, min_y2)

x = np.arange(x1, a1, 0.01)
plt.plot(x, f(x), 'y-', label='Функция убывает')

x = np.arange(a1, a2, 0.01)
plt.plot(x, f(x), 'r-', label='Функция возрастает')

x = np.arange(a2, min_x, 0.01)
plt.plot(x, f(x), 'y-')

x = np.arange(min_x, x2, 0.01)
plt.plot(x, f(x), 'r-')

plt.axhline(0,color='blue')
plt.axvline(0,color='blue')
plt.legend()
plt.grid()
```
