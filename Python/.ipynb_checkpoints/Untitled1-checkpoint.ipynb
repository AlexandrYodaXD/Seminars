```python
# Импортируем необходимые библиотеки
import matplotlib.pyplot as plt
import numpy as np

from sympy import *
from sympy.plotting import plot

from scipy.optimize import fsolve

init_printing()
```

```python
# Задаем переменной x математический смысл переменной x
x = Symbol('x')
```

```python
# Задаем функцию f(x)
f = -12*x**4*sin(cos(x)) - 18*x**3+5*x**2 + 10*x - 30
f
```

<h3>Выведем на экран график функции f(x) на промежутке x [-10, 10]</h3>

```python
plot(f, (x, -10, 10))
```

<h3>Выведем на экран график функции f(x) на промежутке x [-100, 100]</h3>

```python
plot(f, (x, -100, 100))
```

<h3>Выведем на экран график функции f(x) на промежутке x [-1000, 1000]</h3>

```python
plot(f, (x, -1000, 1000))
```
