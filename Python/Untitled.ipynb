Дана функция f(x) = 5x^2 + 10x - 30

1. Определить корни +

2. Найти интервалы, на которых функция возрастает +
 
3. Найти интервалы, на которых функция убывает +

4. Построить график +

5. Вычислить вершину +

6. Определить промежутки, на котором f > 0

7. Определить промежутки, на котором f < 0

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def func(a, b, c, x):
    return a*x**2 + b*x + c
```

```python
a, b, c = 5, 10, -30
x = np.arange(-10, 10, 0.01)
```

```python
def find_sqrt(a, b, c):
    d = b**2 - 4*a*c
    if d < 0:
        return ('Корней нет',)
    elif d == 0:
        x = -b / (2*a)
        return (round(x, 2),)
    else:
        x1 = (-b + d ** 0.5) / (2 * a)
        x2 = (-b - d ** 0.5) / (2 * a)
        return (round(x1, 2), round(x2, 2))
```

```python
sqrts = find_sqrt(a, b, c)
```

```python
min_y = min(func(a,b,c,x))
```

```python
min_x = find_sqrt(a, b, c - min_y)
min_x = min_x[0]
```

```python
pos_down = np.arange(-10, sqrts[1], 0.01)
neg_down = np.arange(sqrts[1], min_x, 0.01)
neg_up = np.arange(min_x, sqrts[0], 0.01)
pos_up = np.arange(sqrts[0], 10, 0.01)

#корни
plt.plot(sqrts[0], 0, 'go', label=f'Корни ({sqrts[0]}, {sqrts[1]})')
plt.plot(sqrts[1], 0, 'go')

#вершина
plt.plot(min_x, min_y, 'rx', label=f'Вершина ({min_x}, {min_y})')

#pos_down
plt.rcParams['lines.linestyle'] = '-'
plt.plot(pos_down, func(a, b, c, pos_down), 'b', label='Убывает > ноля')

#neg_down
plt.rcParams['lines.linestyle'] = '-.'
plt.plot(neg_down, func(a, b, c, neg_down), 'b', label='Убывает < ноля')

#neg_up
plt.plot(neg_up, func(a, b, c, neg_up), 'y', label='Возрастает < ноля')

#pos_up
plt.rcParams['lines.linestyle'] = '-'
plt.plot(pos_up, func(a, b, c, pos_up), 'y', label='Возрастает > ноля')

plt.legend()
plt.grid()
```
